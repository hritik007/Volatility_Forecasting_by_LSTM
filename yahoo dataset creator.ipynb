```python
import pandas as pd
```

```python
# Adding new column "Returns" to our dataset

data['Returns'] = pd.DataFrame(data['Close'] - data['Open'])
```

```python
# deleting extra columns
data = data.drop(columns=['Open','High','Low','Close','Adj Close','Volume'])
```

```python
data.head(15)

# we only need Returns to calculate volatility
```

```python
# to calculate volatility of the whole dataset

volatility_list_till_n = []

sum_till_n = 0.0
mean_till_n = 0.0
varience_till_n = 0.0
volatility_till_n = 0.0
look_back = 20
first_sum = 0

for i in range(look_back-1):
    sum_till_n = sum_till_n + data['Returns'][i]

j = look_back # j = 20
for i in range(len(data)-20):
    # for 20th element (index = 19) : we take sum of 0, 1, 2, 3, .... 19
    # for 21th : we take sum of 1,2, 3, .... 20 (add next value) (remove value 20 steps before)
    
    sum_till_n = sum_till_n + data['Returns'][i+look_back] - data['Returns'][i]
    mean_till_n = sum_till_n/look_back
    varience_till_n = ((data['Returns'][i] - mean_till_n)**2)/look_back
    volatility_till_n = (252**0.5)*(varience_till_n**0.5)
    
    volatility_list_till_n.append(volatility_till_n)
    j = j + 1
        
```

```python
len(volatility_list_till_n) 
# we can see that size is 20 less than the size of dataset (it's obvious)
# hence to make dataframe, we have to drop first 20 entries from the dataset
```

```python
# droping first 20 enteries from the dataset
data = data.iloc[20:]
len(data)

#now you can see that size of dataset(7087) == size of volatility list (7087)
```

```python
# volatility is fluctuation in price
# adding newly created volatility list to our dataframe

data['Volatility'] = volatility_list_till_n
```

```python
# now you can see we now have volatility :)
data.head(20)

# solve non-continous date problem thou :(
```

```python
# print volatility with time

import matplotlib.pyplot as plt

plt.plot(data['Volatility'])
plt.xlabel('time')
plt.ylabel('Volatility')
plt.show()
```
